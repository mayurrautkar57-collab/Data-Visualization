```python
sns.barplot(x=ticker['High'], y=ticker['Low'])
```

```python
# Aggregate 'High' and 'Low' prices by 'Ticker'
ticker_high_low_avg = ticker.groupby('Ticker')[['High', 'Low']].mean()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(ticker_high_low_avg, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title('Average High and Low Prices per Ticker')
plt.xlabel('Price Metric')
plt.ylabel('Ticker')
plt.show()
```

```python
sns.pairplot
```

```python
# @title Select two tickers for volume comparison
selected_ticker1 = 'AAPL' # @param {type:"string"}
selected_ticker2 = 'MSFT' # @param {type:"string"}

# Ensure 'Date' column is datetime type
ticker['Date'] = pd.to_datetime(ticker['Date'])

# Filter data for the selected tickers
compare_df = ticker[ticker['Ticker'].isin([selected_ticker1, selected_ticker2])]

# Plot the volume comparison
plt.figure(figsize=(15, 7))
sns.lineplot(data=compare_df, x='Date', y='Volume', hue='Ticker')
plt.title(f'Share Volume Comparison: {selected_ticker1} vs {selected_ticker2}')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(True)
plt.legend(title='Ticker')
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x=ticker['High'], y=ticker['Low'])
plt.title('High vs Low Stock Prices')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.show()
```

```python
import plotly.express as px

# Ensure ticker_high_low_avg is available (from previous heatmap cell)
# If it's not, you might need to re-run the heatmap cell or define it here.
if 'ticker_high_low_avg' not in locals():
    ticker_high_low_avg = ticker.groupby('Ticker')[['High', 'Low']].mean()

# Create an interactive bar plot for 'High' prices
fig = px.bar(
    ticker_high_low_avg.reset_index(),
    x='Ticker',
    y='High',
    title='Average High Price per Ticker (Interactive)',
    labels={'High': 'Average High Price'},
    hover_data={'Low': True} # Add 'Low' price to hover information
)

fig.show()
```

```python
ticker['Ticker'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('PIe Chart distrbution of ticker')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram for the 'Close' price distribution across all tickers
plt.figure(figsize=(10, 6))
sns.histplot(ticker['Close'], bins=50, kde=True)
plt.title('Distribution of Closing Share Prices (All Tickers)')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# @title Select a ticker to view its 'Close' price distribution

all_tickers = ticker['Ticker'].unique().tolist()
selected_ticker_hist = 'AAPL' # @param ['AAPL', 'AMZN', 'GOOGL', 'META', 'MSFT', 'NVDA'] {type:"category"}

# Filter data for the selected ticker
ticker_filtered = ticker[ticker['Ticker'] == selected_ticker_hist]

# Create a histogram for the 'Close' price of the selected ticker
plt.figure(figsize=(10, 6))
sns.histplot(ticker_filtered['Close'], bins=50, kde=True, color='skyblue')
plt.title(f'Distribution of Closing Share Prices for {selected_ticker_hist}')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
